1(A) Write a python NLTK program to split the text / paragraph from genesis corpus and display it into a list of
words. Remove the punctuation and stop words from the given text and perform stemming

```python
import nltk
nltk.download('all')
```

```python
from nltk.corpus import genesis
nltk.corpus.genesis.fileids()
```

```python
with nltk.corpus.genesis.open('english-kjv.txt') as f:
  data = f.read()

data = data.replace('\n', " ")
data
```

```python
listOfWords = data.split(" ")
listOfWords
print(f"Length of tokens list: {len(listOfWords)}")
```

```python
print(listOfWords)
```

```python
from nltk.corpus import stopwords
stopWords = (stopwords.words('english'))
stopWords
```

```python
# stopword removal
def removeStopwords(listOfWords):
  result = list()
  for w in listOfWords:
    if w.lower() not in stopWords:
      result.append(w)
  return result
listOfWords = removeStopwords(listOfWords)
```

```python
listOfWords
```

```python
print(f"Length of tokens list after removing stop words : {len(listOfWords)}")
text = " ".join(listOfWords)
```

```python
#removing punctuation
tokenizer = nltk.RegexpTokenizer(r"\w+")
newText = tokenizer.tokenize(text)
text = " ".join(newText)
```

```python
print(f"Removing punctuation marks & stopwords: \n\n{text}")
```

```python
data = text.split(" ")
data
```

```python
#Performing stemming & POS Tagging

from nltk.stem import PorterStemmer
ps = PorterStemmer()
for word in data:
  print(f"{word} -> {ps.stem(word)}")
tags = nltk.pos_tag(data)
```

```python
for i in tags:
  print(f"{i[0]} -> {i[1]}")
```

1(B) Write a Python NLTK program to tokenize sentences in languages other than English (german).

```python
with nltk.corpus.genesis.open('german.txt') as f:
  data = f.read()
germanText = data[:910]
germanText = germanText.replace('\n', " ")
germanText
```

```python
#Sentence Tokenization of german txt
from nltk import sent_tokenize
sentenceTokenizer = sent_tokenize(germanText)
sentenceTokenizer
```
